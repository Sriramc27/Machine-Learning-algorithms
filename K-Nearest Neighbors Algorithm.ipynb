```python
# Defining  training and test samples, grouping them into two classes.

training_samples = {
    "Class1": ['X1' , 'X5', 'X9', 'X13', 'X17'],
    "Class2": ['X3' , 'X7' , 'X11' , 'X15' , 'X19']
}

test_samples = {
    "Class1": ['X2', 'X6', 'X10', 'X14', 'X18'],
    "Class2": ['X4', 'X8', 'X12', 'X16']
}
```

```python
# Defining the function to find the K nearest neighbors for training samples
def find_k_nearest_neighbors_data(sample, k):
    sample_distances = distances[sample]
    sorted_samples = sorted(sample_distances.items(), key=lambda x: x[1]) # Sorting distances by ascending order.
    return sorted_samples[:k]  # Return the K nearest neighbors.
```

```python
# Defining the function to calculate the error rate for training samples
def calculate_training_error_rate(k):
    errors = 0
    # Looping through training sample classes.
    for class_name in training_samples: 
        for sample in training_samples[class_name]:  # Loop through samples in each class
            nearest_neighbors = find_k_nearest_neighbors_data(sample, k) #calling find_k_nearest_neighbors_data to Find K nearest neighbors.
            class_counts = {"Class1": 0, "Class2": 0} # Initialize class count dictionary.
            for neighbor in nearest_neighbors: # Loop through nearest neighbors.
                for class_name in training_samples: # Loop through training sample classes.
                    if neighbor[0] in training_samples[class_name]: # Checking which class the neighbor belongs to.
                        # Increment the count for that class.
                        class_counts[class_name] += 1
            predicted_class = max(class_counts.items(), key=lambda x: x[1])[0] # Predicting the class based on counts.
            # Checks if the predicted class is incorrect.
            if predicted_class != class_name:  
                errors += 1  # Increment the error count.
    return errors / sum(len(training_samples[class_name]) for class_name in training_samples)  # Calculates error rate.
```

```python
# Defining the function to calculate the error rate for test samples
def calculate_testing_error_rate(k):
    errors = 0
    for class_name in test_samples: # Looping through test sample classes.
        for sample in test_samples[class_name]: # Loop through samples in each class
            nearest_neighbors = find_k_nearest_neighbors_data(sample, k)  #calling find_k_nearest_neighbors_data to Find K nearest neighbors.
            # Initialize class count dictionary.
            class_counts = {"Class1": 0, "Class2": 0}
            for neighbor in nearest_neighbors:  # Loop through nearest neighbors.
                for class_name in training_samples: # Loop through training sample classes.
                    if neighbor[0] in training_samples[class_name]: # Checking which class the neighbor belongs to.
                        # Increments the count for that class.
                        class_counts[class_name] += 1
            predicted_class = max(class_counts.items(), key=lambda x: x[1])[0] # Predicting the class based on count
            # Checks if the predicted class is incorrect.
            if predicted_class != class_name: 
                errors += 1  # Increments the error count.
    return errors / sum(len(test_samples[class_name]) for class_name in test_samples) # Calculates error rate.
```

```python
# Specifing the K values you want to evaluate
k_values = [1, 3, 5, 7, 9, 10]

# Lists to store error rates for training and testing
training_error_rates = []
testing_error_rates = []
```

```python
# Calculate the training and testing error rates for different values of K
for k in k_values:
    # Calculates training  and testing error rates for K
    training_error_rate = calculate_training_error_rate(k)
    testing_error_rate = calculate_testing_error_rate(k)
    training_error_rates.append(training_error_rate)  # Store training error rate.
    testing_error_rates.append(testing_error_rate)  # Store testing error rate.
    print(f"K={k}, Training Error Rate={training_error_rate}, Testing Error Rate={testing_error_rate}")
```

```python
# Draw the error vs. K plot
import matplotlib.pyplot as plt

# Create a plot to visualize training and test errors for different K values
plt.figure(figsize=(8, 6))
plt.plot(k_values, training_error_rates, marker='o', label='Training Error')  # Plots training error rates.
plt.plot(k_values, testing_error_rates, marker='o', label='Test Error')  # Plots testing error rates.
plt.title('Training and Test Errors vs. K')
plt.xlabel('K (Number of Neighbors)')
plt.ylabel('Error Rate')
plt.legend() # Display a legend.
plt.grid(True)
plt.show()
```
